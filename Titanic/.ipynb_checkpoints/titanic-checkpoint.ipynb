```python
# 根据事故背景，首先分析下直观上对预测有较大影响的特征
survived0 = data_train.Pclass[data_train.Survived == 0].value_counts()
survived1 = data_train.Pclass[data_train.Survived == 1].value_counts()
df = DataFrame({'unsurvived': survived0, 'survived': survived1})
df.plot(kind='bar', stacked=True)
plt.title('class to survived')
plt.xlabel('Pclass')
plt.ylabel('number')
plt.show()
```

```python
malePro = data_train.Survived[data_train.Sex == 'male'].value_counts()
femalePro = data_train.Survived[data_train.Sex == 'female'].value_counts()
df = DataFrame({'male': malePro, 'female': femalePro})
df.plot(kind='bar', stacked=True)
plt.title('the probabily of sex to survived')
plt.xlabel('sex')
plt.ylabel('number')
plt.show()
```

```python
# 2、粗略分析数据集的特征，去除和测试结果无关的特征，在这里ticket无法提出有效信息，与测试结果无关
data_train = data_train.drop(['Ticket'], axis=1)
data_test = data_test.drop(['Ticket'], axis=1)
```

```python
# 3、分析数据集中的缺失值信息，并处理缺失值
total = data_train.isnull().sum().sort_values(ascending=False)
percent = (data_train.isnull().sum() / data_train.isnull().count()).sort_values(ascending=False)
df = pd.concat([total, percent], axis=1)
df.head()  # 可以考虑删除Embarked为Nan的两个数据
```

```python
# 分析Cabin中缺失值的数据集和未缺失的数据集,可以明显发现Cabin的存活率明显要低
isNull = data_train[data_train.Cabin.isnull()]['Survived'].mean()
notNull = data_train[data_train.Cabin.notnull()]['Survived'].mean()
print("isnull mean: {}".format(isNull))
print("notnull mean: {}".format(notNull))
```

```python
# 将Cabin中的缺失值标为0，未缺失标为1
data_train['hasCabin'] = data_train['Cabin'].apply(lambda x: 0 if type(x) == float else 1)
data_test['hasCabin'] = data_test['Cabin'].apply(lambda x: 0 if type(x) == float else 1)
data_train.head()
```

```python
# 去除原来的Cabin属性
data_train = data_train.drop(["Cabin"], axis=1)
data_test = data_test.drop(["Cabin"], axis=1)
data_train.head()
```

```python
# 将性别转换成数值的形式
data_train['Sex'] = data_train['Sex'].map({'female': 1, 'male': 0}).astype(int)
data_test['Sex'] = data_test['Sex'].map({'female': 1, 'male': 0}).astype(int)
data_train.head()
```

```python
# 对Age中的缺失值采用中位数来补齐，并且基于sex和pclass两个属性来分段去中位数补齐
def fillAge(dataSet):
    guess_ages = np.zeros((2, 3))

    for i in range(2):
        for j in range(3):
            guess_df = dataSet[(dataSet['Sex'] == i) & (dataSet['Pclass'] == j + 1)]['Age'].dropna()
            
            guess_ages[i, j] = int(guess_df.median() / 0.5 + 0.5) * 0.5
    for i in range(2):
        for j in range(3):
            dataSet.loc[(dataSet['Sex'] == i) & (dataSet['Pclass'] == j + 1) & (dataSet.Age.isnull()), 'Age'] = guess_ages[i, j]
    dataSet['Age'] = dataSet['Age'].astype(int)
    
    return dataSet

data_train = fillAge(data_train)
data_test = fillAge(data_test)
data_train.head()
```

```python
# 查看年龄的密度曲线
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.figure()
data_train['Age'].plot(kind='kde')
survived0 = data_train.Age[data_train['Survived'] == 0].plot(kind='kde')
survived1 = data_train.Age[data_train['Survived'] == 1].plot(kind='kde')

plt.title('age distributed respect for survived')
plt.ylabel('dense')
plt.legend(('total', 'unsurvived', 'survived'), loc='best')
plt.show()
```

```python
# 对年龄做分段处理, 因为不同年龄段之间的幸存率是不一样的
def ageTransfrom(dataSet):
    dataSet.loc[dataSet['Age'] <= 16, 'Age'] = 0
    dataSet.loc[(dataSet['Age'] > 16) & (dataSet['Age'] <= 32), 'Age'] = 1
    dataSet.loc[(dataSet['Age'] > 32) & (dataSet['Age'] <= 48), 'Age'] = 2
    dataSet.loc[(dataSet['Age'] > 48) & (dataSet['Age'] <= 64), 'Age'] = 3
    dataSet.loc[dataSet['Age'] > 64, 'Age'] = 4
    return dataSet

data_train = ageTransfrom(data_train)
data_test = ageTransfrom(data_test)
data_train.head()
```

```python
# 4、从现有特征中提取新特征，在这里name属性中可以提取出描述身份的信息，例如Mr
def fromNameExtractInfo(dataSet):
    
    dataSet['Title'] = dataSet.Name.str.extract(r'([A-Za-z]+)\.', expand=False)
    dataSet['Title'] = dataSet['Title'].replace(['Lady', 'Countess', 'Capt', 'Col', 'Don', 'Dr', 'Major', 'Rev',
                                                 'Sir', 'Jonkheer', 'Dona'], 'Rare')
    dataSet['Title'] = dataSet['Title'].replace('Mlle', 'Miss')
    dataSet['Title'] = dataSet['Title'].replace('Ms', 'Miss')
    dataSet['Title'] = dataSet['Title'].replace('Mme', 'Mrs')

    return dataSet
data_train = fromNameExtractInfo(data_train)
data_test = fromNameExtractInfo(data_test)
data_train.head()
```

```python
# 分析提取出来的新特征与预测值之间的关系
fig = plt.figure()

plt.subplot2grid((2, 3), (0, 0))
data_train.Survived[data_train.Title == 'Mr'].value_counts().plot(kind='bar')
plt.xlabel('Mr')  # 活下来的概率小

plt.subplot2grid((2, 3), (0, 1))
data_train.Survived[data_train.Title == 'Miss'].value_counts().plot(kind='bar')
plt.xlabel('Miss')  # 活下来的概率大

plt.subplot2grid((2, 3), (1, 0))
data_train.Survived[data_train.Title == 'Mrs'].value_counts().plot(kind='bar')
plt.xlabel('Mrs')  # 活下来的概率更大

plt.subplot2grid((2, 3), (1, 1))
data_train.Survived[data_train.Title == 'Master'].value_counts().plot(kind='bar')
plt.xlabel('Master')  # 活下来的概率比死亡的概率大

plt.subplot2grid((2, 3), (1, 2))
data_train.Survived[data_train.Title.isin(['Mr', 'Miss', 'Mrs', 'Master']) == False].value_counts().plot(kind='bar')
plt.xlabel('other')  # 死亡的概率大于活下来的概率

plt.show()
```

```python
# 不同的身份存活率不一样，我们根据存活率大小进行数值化
def titleNumber(dataSet):
    
    title_mapping = {'Mr': 1, 'Miss': 4, 'Mrs': 5, 'Master': 3, 'Rare': 2}

    dataSet['Title'] = dataSet['Title'].map(title_mapping)
    dataSet['Title'] = dataSet['Title'].fillna(0)

    dataSet = dataSet.drop(['Name'], axis=1)

    return dataSet

data_train = titleNumber(data_train)
data_test = titleNumber(data_test)
data_train.head()
```

```python
# 5、合并特征，构建出新的特征，Parch和SibSp有点类似，可以考虑合并
combine1 = data_train[(data_train['Parch'] == 0) & (data_train['SibSp'] == 0)]['Survived'].value_counts()
combine2 = data_train[(data_train['Parch'] == 0) & (data_train['SibSp'] > 0)]['Survived'].value_counts()
combine3 = data_train[(data_train['Parch'] > 0) & (data_train['SibSp'] == 0)]['Survived'].value_counts()
combine4 = data_train[(data_train['Parch'] > 0) & (data_train['SibSp'] > 0)]['Survived'].value_counts()
print("combine1: {}".format(combine1))
print("combine2: {}".format(combine2))
print("combine3: {}".format(combine3))
print("combine4: {}".format(combine4))
```

```python
def extractSibSpAndParch(dataSet):
    
    dataSet['FamilySize'] = dataSet['SibSp'] + dataSet['Parch'] + 1
    dataSet['IsAlone'] = 0
    dataSet.loc[dataSet['FamilySize'] == 1, 'IsAlone'] = 1
    dataSet = dataSet.drop(['Parch', 'SibSp'], axis=1)

    return dataSet

data_train = extractSibSpAndParch(data_train)
data_test = extractSibSpAndParch(data_test)
data_train.head()
```

```python
# 6、分析类目型特征Embarked，不同的登船口岸存活率
fig = plt.figure()
data_train.Embarked.value_counts().plot(kind='bar')
plt.title(u'number of embarked')
plt.ylabel(u'number')
plt.show()
```

```python
fig = plt.figure()
survived0 = data_train.Embarked[data_train.Survived == 0].value_counts()
survived1 = data_train.Embarked[data_train.Survived == 1].value_counts()
df = DataFrame({'survived': survived1, 'unsurvived': survived0})
df.plot(kind='bar', stacked=True)
plt.title('the survived probably respect for Embarked')
plt.xlabel('Embarked')
plt.ylabel('number')
plt.show()
```

```python
def embarkedTransform(dataSet):

    freq_port = dataSet.Embarked.dropna().mode()[0]
    dataSet.loc[dataSet.Embarked.isnull(), 'Embarked'] = freq_port

    dataSet['Embarked'] = dataSet['Embarked'].map({'S': 0, 'Q': 1, 'C': 2}).astype(int)

    return dataSet

data_train = embarkedTransform(data_train)
data_test = embarkedTransform(data_test)
data_train.head()
```

```python
def fareTransform(dataSet):

    dataSet.loc[dataSet['Fare'] <= 7.91, 'Fare'] = 0
    dataSet.loc[(dataSet['Fare'] > 7.91) & (dataSet['Fare'] <= 14.454), 'Fare'] = 1
    dataSet.loc[(dataSet['Fare'] > 14.454) & (dataSet['Fare'] <= 31), 'Fare'] = 2
    dataSet.loc[dataSet['Fare'] > 31, 'Fare'] = 3
    dataSet['Fare'] = dataSet['Fare'].astype(int)

    return dataSet
data_train = fareTransform(data_train)
data_test = fareTransform(data_test)
data_train.head()
```
