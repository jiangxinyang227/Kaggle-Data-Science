```python
data_train = pd.read_csv('./data/train.csv')
data_test = pd.read_csv('./data/test.csv')
combine = [data_train, data_test]
data_train.head()
```

```python
# 删除Id
trainId = data_train['Id']
testId = data_test['Id']
data_train.drop(['Id'], axis=1, inplace=True)
data_test.drop(['Id'], axis=1, inplace=True)
```

```python
# 删除异常点
data_train = data_train.drop(data_train[(data_train['GrLivArea'] > 4000) & (data_train['SalePrice'] < 300000)].index)
```

```python
ntrain = data_train.shape[0]
trainY = data_train.SalePrice.values
dataSet = pd.concat([data_train, data_test], axis=0)
dataSet.drop(['SalePrice'], axis=1, inplace=True)
dataSet.shape
```

```python
total = dataSet.isnull().sum().sort_values(ascending=False)
total.head(40)
```

```python
dataSet['GarageYrBlt'] = dataSet['GarageYrBlt'].fillna(method='ffill')
dataSet['GarageYrBlt'].isnull().sum()
```

```python
garage = list(dataSet['GarageYrBlt'])
built = list(dataSet['YearBuilt'])
error = 0
for i in range(len(built)):
    if built[i] != garage[i]:
        error +=1
        
print(error / len(built))
    
```
