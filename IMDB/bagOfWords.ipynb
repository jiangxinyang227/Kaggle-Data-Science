```python
import re
import pandas as pd
from bs4 import BeautifulSoup
import nltk
```

```python
train = pd.read_csv('./data/labeledTrainData.tsv', header=0, delimiter='\t', quoting=3)
test = pd.read_csv('./data/testData.tsv', header=0, delimiter='\t', quoting=3)
```

```python
print("Train shape: {}".format(train.shape))
print("Test shape: {}".format(test.shape))
```

```python
print('Train columns: {}'.format(train.columns))
print("Test columns: {}".format(test.columns))
```

```python
train.head()
```

```python
test.head()
```

```python
train['review'][0]
```

```python
def removeHtmlTag(review):
    beau = BeautifulSoup(review)
    return beau.get_text()

train['review'] = train['review'].apply(removeHtmlTag)
test['review'] = test['review'].apply(removeHtmlTag)
train.head()
```

```python
train['review'][0]
```

```python
def removeSpecialChar(review):
    review = re.sub("[^a-zA-Z]", " ", review)
    return review

train['review'] = train['review'].apply(removeSpecialChar)
test['review'] = test['review'].apply(removeSpecialChar)
train.head()
```

```python
train['review'][0]
```

```python
train['review'] = train['review'].apply(lambda x: x.lower())
test['review'] = test['review'].apply(lambda x: x.lower())
```

```python
train['review'][0]
```

```python
train['review'] = train['review'].apply(lambda x: x.split(" "))
test['review'] = test['review'].apply(lambda x: x.split(" "))
train.head()
```

```python
train['review'][0]
```

```python
def removeBlank(review):
    review = [word for word in review if word]
    return review

train['review'] = train['review'].apply(removeBlank)
test['review'] = test['review'].apply(removeBlank)
train['review'][0]
```

```python
with open('./data/english') as f:
    stopwords = f.read()
    stopwords = stopwords.splitlines()
stopwords
```

```python
def removeStopWords(review):
    review = [word for word in review if word not in stopwords]
    return review

train['review'] = train['review'].apply(removeStopWords)
test['review'] = test['review'].apply(removeStopWords)
train['review'][0]
```

```python
train['review'] = train['review'].apply(lambda x: " ".join(x))
test['review'] = train['review'].apply(lambda x: " ".join(x))
train['review'][0]
```

```python
train.head()
```

```python
test.head()
```

```python
train_data = train['review']
labels = train['sentiment']
test_data = test['review']
```

```python
trainList = train_data.tolist()
testList = test_data.tolist()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(analyzer='word', max_features = 5000)
train_features = vectorizer.fit_transform(trainList)
train_features = train_features.toarray()
test_features = vectorizer.transform(testList)
test_features = test_features.toarray()

print("Train shape: {}".format(train_features.shape))
print("Test shape: {}".format(test_features.shape))
```

```python
featureName = vectorizer.get_feature_names()
featureName
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=200, max_depth=10, min_samples_split=4)
classifier.fit(train_features, labels)
result = classifier.predict(test_features)
output = pd.DataFrame(data={"id": test['id'], "sentiment": result})
output.to_csv("./data/bagOfWords.csv", index=False, quoting=3)
```
